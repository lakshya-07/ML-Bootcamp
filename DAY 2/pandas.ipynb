```python
#loc-loc attribute allows indexing and slicing that reference the explicit index(Actual index given to value)
series_obj=pd.Series(['Amazon Web Services (AWS)', 'Microsoft Azure', 'Google Cloud', 'Firebase'],index=[1,3,2,5])
print("Accessing the second element of the series using explicit index and loc attribute:",series_obj.loc[3])
print("Getting the slice of first three elements of the series using explicit indexing and loc attribute:")
print(series_obj.loc[:2])#here in loc the final slice is included as well
```

```python
#iloc-iloc attribute allows indexing and slicing that reference the implicit index(Numerical index given to value)
series_obj=pd.Series(['Amazon Web Services (AWS)', 'Microsoft Azure', 'Google Cloud', 'Firebase'],index=[1,3,2,5])
print("Accessing the second element of the series using implicit index and iloc attribute:",series_obj.iloc[1])
print("Getting the slice of first three elements of the series using implicit indexing and iloc attribute:")
print(series_obj.iloc[:3])#here in iloc the final slice is not included
```

```python
# Define a function to categorize the 'PERCENTAGE_OF_VOTES'
def categorize_percentage(total_votes, table):
    if total_votes > np.mean(table['TOTAL_VOTES']):
        return 'ABOVE AVERAGE'
    else:
        return 'BELOW AVERAGE'

# Apply the function to create a new column 'VOTE_CATEGORY'
cloud_survey_table['VOTE_CATEGORY'] = cloud_survey_table['TOTAL_VOTES'].apply(categorize_percentage, table=cloud_survey_table)

# Display the updated DataFrame
cloud_survey_table
```

```python
# Define a transform function to append "%" to the percentages
def transform_percentages(column):
    return column.astype(str) + '%'

# Apply the transform function to 'PERCENTAGE_OF_VOTES'
cloud_survey_table['PERCENTAGE_TRANSFORMED'] = cloud_survey_table.groupby('CATEGORY')['PERCENTAGE_OF_VOTES'].transform(transform_percentages)

# Display the updated DataFrame
cloud_survey_table
```
