# MatplotLib

##### Matplotlib is a versatile data visualization library in Python, initially developed in 2002 by John Hunter and is inspired by the MATLAB plotting.It was adopted by the Space Telescope Science Institute. Its cross-platform compatibility and support for various backends and output types have contributed to its widespread use in the scientific Python community.

##### Although there are other tools ggplot in R and d3js which are much better in  visualization,Matplotlib remains a robust, well-tested, and cross-platform graphics engine. We have APIs that ensure a much more cleaner user interface such as Seaborn

```python
#plotting some trigonometric points
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x),"--")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming you have a DataFrame named stock_prices with the provided data
# If not, you can create it using: stock_prices = pd.read_csv("your_data.csv")

# Convert the "Date" column to datetime
stock_prices["Date"] = pd.to_datetime(stock_prices["Date"])

# Plotting a line plot for Microsoft stock prices over time
plt.figure(figsize=(12, 6))
plt.plot(stock_prices["Date"], stock_prices["Microsoft"], label="Microsoft", marker='o')
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.title("Microsoft Stock Prices Over Time")
plt.legend()
plt.show()
```

```python
stock_prices.head().melt(id_vars="Date", var_name="Company", value_name="Stock Price")
```

```python
plt.figure(figsize=(10, 6))
latest_prices=stock_prices.head().melt(id_vars="Date", var_name="Company", value_name="Stock Price")
plt.bar(latest_prices["Company"], latest_prices["Stock Price"])
plt.xlabel("Company")
plt.ylabel("Stock Price (USD)")
plt.title("Latest Stock Prices")
plt.show()
```

```python
# Plotting a Histogram for Microsoft stock prices
plt.figure(figsize=(10, 6))
plt.hist(stock_prices["Microsoft"], bins=20, color='skyblue', edgecolor='black')
plt.xlabel("Microsoft Stock Price")
plt.ylabel("Frequency")
plt.title("Histogram of Microsoft Stock Prices")
plt.show()
```

Positive Correlation: If the points generally form an upward-sloping pattern, it suggests a positive correlation. In other words, when Microsoft stock prices increase, Google stock prices tend to increase as well.

Negative Correlation: Conversely, if the points form a downward-sloping pattern, it suggests a negative correlation. This means that as Microsoft stock prices go up, Google stock prices tend to go down, and vice versa.

No Correlation: If the points are scattered without a clear pattern, it suggests no correlation. Changes in Microsoft stock prices do not seem to be related to changes in Google stock prices.

Strength of Correlation: The tighter the cluster of points, the stronger the correlation. If the points are spread out, the correlation may be weaker.

Outliers: Identify any points that do not follow the general pattern. Outliers may indicate unusual events or factors affecting one of the stocks.

```python
import matplotlib.pyplot as plt

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(latest_prices[latest_prices["Company"] == "Microsoft"]["Stock Price"],
            latest_prices[latest_prices["Company"] == "Google"]["Stock Price"],
            c='blue', label='Microsoft vs. Google', alpha=0.7)

# Customize the plot
plt.xlabel("Microsoft Stock Price (USD)")
plt.ylabel("Google Stock Price (USD)")
plt.title("Scatter Plot: Microsoft vs. Google Stock Prices")
plt.legend()

# Show the plot
plt.show()
```
